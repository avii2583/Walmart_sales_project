# Business problems

```python
# Q1 - : What are the different payment methods, and how many transactions and
# items were sold with each method?

query = """ select  payment_method , COUNT(*) as no_payments,SUM(quantity) as no_qty_sold
from walmart 
GROUP BY payment_method
ORDER BY no_payments DESC
"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['payment_method','no_payments','no_qty_sold'])
dg
```

```python
# Q2 - Which category received the highest average rating in each branch?

query = """ SELECT * FROM 
(select branch,category , AVG(rating) as avg_rating ,
RANK() OVER(PARTITION BY branch ORDER BY AVG(rating) DESC)  as rankk
from walmart
GROUP BY branch,category)
AS t
WHERE rankk  = 1

"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['branch','category','highest_avg_rating','rank'])
dg
```

```python
# Q3 - What is the busiest day of the week for each branch based on transaction
# volume?

query = """ SELECT branch,
DAYNAME(STR_TO_DATE(`date`, '%d/%m/%y')) AS day_of_week,
COUNT(*) AS transactions
FROM walmart
GROUP BY branch, day_of_week
ORDER BY branch, transactions DESC;


"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['branch','day_of_week','no_of_transactions'])
dg
```

```python
# Q4 - How many items were sold through each payment method?

query = """ 
SELECT payment_method,
SUM(quantity) AS items_sold               
FROM walmart
GROUP BY payment_method;


"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['payment_method','no_qty_sold'])
dg
```

```python
# Q5 - What are the average, minimum, and maximum ratings for each category in
# each city?

query = """ 
SELECT City,category,
AVG(rating) AS avg_rating,
MIN(rating) AS min_rating,
MAX(rating) AS max_rating
FROM walmart
GROUP BY City, category;

"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['city','category','avg_rating','min_rating','max_rating'])
dg
```

```python
# Q6 - What is the total profit for each category, ranked from highest to lowest?

query = """ 
SELECT category,
SUM(total * profit_margin) AS total_profit,
SUM(total) AS total_revenue
FROM walmart
GROUP BY category
ORDER BY total_profit DESC;

"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['category','total_profit','total_revenue'])
dg
```

```python
# Q7 - What is the most frequently used payment method in each branch?

query = """ 
SELECT branch,payment_method,
COUNT(*) AS usage_count,
RANK() OVER 
(PARTITION BY branch
ORDER BY COUNT(*) DESC) AS method_rank
FROM walmart
GROUP BY branch, payment_method
ORDER BY branch, method_rank;
"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['branch','payment_method','no_of_payments','rank'])
dg
```

```python
# Q8 - How many transactions occur in each shift (Morning, Afternoon, Evening)
# across branches?

query = """ 
SELECT branch,shift,
COUNT(*) AS transactions
FROM (SELECT branch,CASE
            WHEN TIME(time) >= '06:00:00' AND TIME(time) < '12:00:00'
                 THEN 'Morning'
            WHEN TIME(time) >= '12:00:00' AND TIME(time) < '18:00:00'
                 THEN 'Afternoon'
            ELSE 'Evening'
        END AS shift
    FROM walmart) AS t
GROUP BY branch, shift
ORDER BY branch, shift;

"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['branch','shift','no_of_transactions'])
dg
```

```python
# Q9 - Which branches experienced the largest decrease in revenue compared to
# the previous year?

query = """ 
WITH yearly_revenue AS 
(SELECT Branch,
YEAR(STR_TO_DATE(date, '%d/%m/%y')) AS year,
SUM(total) AS revenue
FROM walmart
GROUP BY Branch, year),
revenue_comparison AS 
(SELECT Branch, year AS current_year,
revenue AS current_year_revenue,
LAG(revenue) OVER (PARTITION BY Branch ORDER BY year) AS last_year_revenue
FROM yearly_revenue)
SELECT Branch, last_year_revenue, current_year_revenue,
ROUND(100.0 * (last_year_revenue - current_year_revenue) / last_year_revenue,2) AS percentage_decrease
FROM revenue_comparison
WHERE last_year_revenue IS NOT NULL
AND current_year_revenue < last_year_revenue
ORDER BY percentage_decrease DESC;

"""

cur.execute(query)

data = cur.fetchall()
dg = pd.DataFrame(data,columns=['branch','last_year_rev','current_year_rev','percentage_dec'])
dg
```
